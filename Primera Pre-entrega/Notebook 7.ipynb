9) ¿Hay valores faltantes en alguna de las columnas de `expvida`?

```python
# Obtengo la cantidad de datos faltantes en cada columna
datos_faltantes = expvida.isnull().sum() #sumar la cantidad de datos nulos por columna

print("Cantidad de datos faltantes por columna:")
print(datos_faltantes)
```

10) Selecciona el maximo de datos faltantes. Calcula que porcentaje de los datos representan los datos faltantes.

```python
# Obtengo el maximo valor de datos faltantes y lo guardo en la variable masnan
masnan = expvida.isnull().sum().max()

print("Valor maximo de datos faltantes:" , masnan)
```

```python
# Obtengo la cantidad de filas total y divido el maximo valor de datos faltantes por este numero
# y lo multiplico por 100
cantidad_total_filas = expvida.shape[0] #Usar expvida.shape[0] o len(expvida) para llamar al total de filas
print("La cantidad total de filas en el DataFrame es:", cantidad_total_filas)

division_filas_masnan = cantidad_total_filas / masnan

multiplicado_por_100 = division_filas_masnan * 100
print("Total de filas dividido al max valor de datos faltantes multiplicado por 100: ", multiplicado_por_100)
```

```python
# Imprimo el maximo porcentaje de datos faltantes. Uso f-strings.

# Datos faltantes en cada columna
porcentaje_datos_faltantes = (datos_faltantes / cantidad_total_filas) *100
print("Porcentaje de datos faltantes por columna:")
print (porcentaje_datos_faltantes)

# Máximo porcentaje de datos faltantes
maximo_porcentaje_faltante = porcentaje_datos_faltantes.max()
print(f"Máximo porcentaje de datos faltantes: {maximo_porcentaje_faltante}")
```

12) ¿Cual es el promedio de expectativa de vida (Life expectancy) en general?

```python
# Obtengo el promedio de la columna
promedio_life_expectancy = expvida['life_expectancy'].mean()

print(f"Promedio de la columna Life expectancy: {promedio_life_expectancy}")
```

13) ¿Cual es el promedio de expectativa de vida (Life expectancy) por país?

```python
# Obtengo el promedio por pais
# groupby = agrupar por
promedio_life_pais = expvida.groupby('Country')['life_expectancy'].mean()

print("Promedio de la columna Life expectancy por país:")
print(promedio_life_pais)
```

14) ¿Cual es la expectativa de vida mas alta y mas baja? Usa la funcion de pandas y la funcion de NumPy y compara el resultado. ¿Que paises tienen esas expectativas de vida?

```python
# Obtener valores de la columna de Life expectancy como un array NumPy
life_expectancy_array = expvida['life_expectancy'].values
# Todos los valores en un array
print("Todos los valores de expectativa de vida segun numpy:", life_expectancy_array) 
```

```python
# Maxima expectativa de vida usando numpy
# Máximo de la expectativa de vida sacando aquellos datos NaN
max_life_expectancy = np.nanmax(life_expectancy_array)
```

```python
# Maxima expectativa de vida usando pandas
max_life_expectancy_pandas = expvida['life_expectancy'].max()
```

```python
# Minima expectativa de vida usando numpy
min_life_expectancy = np.nanmin(life_expectancy_array)
```

```python
# Minima expectativa de vida usando pandas
min_life_expectancy_pandas = expvida['life_expectancy'].min()
```

```python
# Muestro los resultados maximos
print("Máxima expectativa de vida segun numpy:", max_life_expectancy)

print("Máxima expectativa de vida segun pandas:", max_life_expectancy_pandas)
```

```python
# Muestro los resultados minimos
print("Minima expectativa de vida segun numpy:", min_life_expectancy)

print("Minima expectativa de vida segun pandas:", min_life_expectancy_pandas)
```

```python
# Filtro las columnas para mostrar solo las filas que tengan valor
# igual al maximo

#loc es para filtrar
filas_max_expectativa = expvida.loc[expvida['life_expectancy'] == max_life_expectancy]

columnas_max_expectativa = filas_max_expectativa['life_expectancy']

print("Filas que tengan valor igual al máximo de expectativa de vida:")
print (columnas_max_expectativa)

```

```python
# Filtro las columnas para mostrar solo las filas que tengan valor
# igual al minimo

#loc es para filtrar
filas_min_expectativa = expvida.loc[expvida['life_expectancy'] == min_life_expectancy_pandas]

columnas_min_expectativa = filas_min_expectativa['life_expectancy']

print("Filas que tengan valor igual al minimo de expectativa de vida:")
print (columnas_min_expectativa)
```

15) ¿Cuantas categorias encontramos en la columna "Status"? _**Pista**_: Lee acerca de la función [unique()](http://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.unique.html?highlight=unique#pandas.Series.unique)

```python
# Obtengo los valores unicos de la columna status
categoria_status = expvida['Status'].unique()

cantidad_categorias = len(categoria_status)

print("Cateogias de la columna Status:", categoria_status)
print("Cantidad de Categorias de la columna Status:", cantidad_categorias)
```

16) ¿Cuantos paises se ubican en cada una de esas categorias de "Status"? _**Pista**_: Lee acerca de la función [value_counts()](http://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.value_counts.html)

```python
# Cuento cuantos valores hay de cada categoria
paises_developing = expvida[expvida['Status'] == 'Developing']
paises_developed = expvida[expvida['Status'] == 'Developed']

count_developing = paises_developing['Status'].value_counts()
count_developed = paises_developed['Status'].value_counts()

# Mostrar los resultados
print("Cantidad de países en la categoría 'Developing':")
print(count_developing)

print("\nCantidad de países en la categoría 'Developed':")
print(count_developed)
```

18) ¿Cuantos paises presentaron algun caso de Sarampion (**Measles**) en el 2015? **_Pista_**: Lee acerca de la función [.count()](http://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.count.html?highlight=count#pandas.DataFrame.count)

```python
# Cuento cuantas filas tienen valores distintos de cero 
casos_de_sarampion_2015 = expvida[(expvida['measles'].notnull()) & (expvida['Year'] == 2015)] 

cantidad_paises_de_sarampion_2015 = casos_de_sarampion_2015['Country'].count()

print("Cantidad de países que presentaron casos de sarampión en 2015:", cantidad_paises_de_sarampion_2015)
```
